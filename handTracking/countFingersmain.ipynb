```python
import os
import cv2
import time
import ipynb.fs.full.HandTrackingModule as htm
```

```python
cap=cv2.VideoCapture(0)
wCam=1280
hCam=1280
path=r"C:\Users\ayush\OneDrive\Desktop\handCountImg"
myList=os.listdir(path) # to get all the name of the elements present in that directory
overLayed=[]
for i in myList:
    imgap=cv2.imread(f'{path}\\{i}') # read and get coordinates of all the img
   # print(imgap.shape) # to get the shape of all the img height and width
    overLayed.append(imgap)
#print(myList)
#cap.set(3,wCam)
#cap.set(4,hCam)
cTime=0
pTime=0
# hand tracking module
detector=htm.handDetector()

while True:
    success,img=cap.read()
     # get the height and width of img and put that img in main img
    
    # hand detector
    img=detector.findHands(img)
    lmList=detector.findPosition(img)
    if len(lmList)!=0:
        #print(lmList)
        lm=[4,8,12,16,20]
        count=0
        for id in range(1,5):
            if(lmList[lm[id]][2]<lmList[lm[id]-2][2]):
                count=count+1
        if(lmList[4][1]>lmList[6][1]):
                count=count+1
        print(lmList)
        # print(count)
        h,w,c=overLayed[count].shape
        img[0:h,0:w]=overLayed[count]
       #         print("Index finger is open")
    
    # fps
    cTime=time.time()
    fps=1/(cTime-pTime)
    pTime=cTime
    cv2.putText(img,f'FPS:{int(fps)}',(420,70),cv2.FONT_HERSHEY_PLAIN,3,(255,0,255),5)
    cv2.imshow("img",img)
    if cv2.waitKey(1) & 0xFF==ord('a'):
        break
cap.release()
cv2.destroyAllWindows()
```
